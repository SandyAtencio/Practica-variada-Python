**Metodos de Cadena**


```
place = '**Pool HOUSE**'

print(place.upper()) # convierte la cadena a mayuscula
print(place.lower()) # convierte la cadena a minuscula
print(place.count('o')) # cuenta cuantas 'o' encuentra en la cadena
print(place.replace('E', 'ER')) # permite remplazar un valor de la cadena
print(place.strip('*')) # elimina caracteres del lao izq y der de la cadena
```

**Metodos de listas**
<p>Aqui explicaremos para que sirve cada metodo de las listas ( el objecto *list*), tambien sirve para tuplas</p>

```
ciudades = ['Bogota', 'Valledupar', 'Bucaramanga']

# metodos de agregacion
ciudades.append('Neiva') # agregamos un elemento al final de la lista
print(ciudades)
ciudades.extend(['Cali', 'Barranquilla']) # agrega varios elementos al final de la lista
print(ciudades)
ciudades.insert(0, 'Medellín') # agrega un elemento en el indice que le indiquemos --> aqui le indicamos en el indice 0
print(ciudades)

# metodos de eliminacion
print(ciudades.pop(-1)) # elimina un elemento por su indice
print(ciudades)
ciudades.remove('Bogota') # elimina un elemento por su valor
print(ciudades)

# metodos de ordenamiento
ciudades.reverse() # invierte el orden de la lista
print(ciudades)
ciudades.sort() # ordena la lista de forma ascendente
print(ciudades)

# metodos de busqueda
print(ciudades.count('Cali')) # imprime cuantas veces aparece la palabra cali en la lista
print(ciudades.index('Neiva')) # imprime el indice donde se encuentra ubicado
```

**<h3>Numpy</h3>**
<p>Numpy es una libreria que nos permite crear matrices y manipularlas</p>

*   las matrices de numpy no pueden contener elementos de diferentes tipos
*   los operadores aritméticos típicos, tales como +, -, * y / tienen un significado diferente para las listas (list) regulares de Python a diferenica de una matriz de numpy.

```
import numpy as np # importamos la libreria de numpy y le asigamos un alias de np
listaNum = [2,4,6,8]
np_listaNum = np.array(listaNum) # convertimos la lista de numeros a una matriz de numpy
```

**Seleccion de elementos de una matriz numpy**

```
np_2d = np.array([['Pera', 'Manzana', 'Uva'], ['Uva', 'Fresa', 'Mango']]) # creamos una matriz 2D de numpy
print(np_2d)
print(np_listaNum[1]) # podemos acceder al elemento 1 de la matriz unidimensional
print(np_listaNum[0:2]) # podemos acceder desde el elemento 0 al elemento 2 (el elemento 2 es excluyente)
print(np_2d[:,0]) # el : indica fila y el 0 indica columna. entonces, quiero traerme todas las filas con la columna 0
np_2d.shape # muestra la forma de la matriz ( # filas, # colum)
```

**Operaciones con matrices de numpy**

**Ejemplo**

`np.array([True, 1, 2]) + np.array([3, 4, False])`

Aqui pretendemos **"CONCATENAR**" dos matricez de numpy, pero... lo que en realidad se ejecuta es una suma de matricez, y la realiza de la siguiente forma:

`np.array([True, 1, 2])`<br>
+<br>
`np.array([3, 4, False])`<br>
=  `array(([4,5,3])` ---> **True equivale a 1 y False equivale a 0 ( como siempre nos han enseñado en el mundo de la programacion :) ) por eso da ese resultado**

```
print(np_listaNum > 3) # preguntamos si cada elemento de la matriz en mayor a 3
print(np_listaNum * 3) # multiplicamos cada elemento de la matriz por 3
print(np_listaNum + np.array([1, 1, 1, 1])) # suma los elementos de la primera matriz mas los elementos de la nueva matriz
```
